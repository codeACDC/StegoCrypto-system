```python
from PIL import Image, ImageDraw
from random import randint
from re import findall
import os

def crypt(message, choice):
    cryp_eng_arr = {'a': 'W', 'b': 'Z', 'c': '2', 'd': '3', 'e': 'S', 'f': '?',\
                    'g': 'B', 'h': 'J', 'i': '"', 'j': 'X', 'k': 'H', 'l': '8',\
                    'm': '9', 'n': 'L', 'o': '<', 'p': 'b', 'q': '$', 'r': 'I',\
                    's': ',', 't': 'T', 'u': 'a', 'v': '4', 'w': '1', 'x': '5',\
                    'y': '(', 'z': ';', 'A': 'E', 'B': '#', 'C': '7', 'D': 'A',\
                    'E': ':', 'F': 'O', 'G': '^', 'H': ')', 'I': '@', 'J': "'",\
                    'K': 'R', 'L': 'M', 'M': '=', 'N': 'C', 'O': '%', 'P': ']',\
                    'Q': '+', 'R': '-', 'S': 'U', 'T': 'V', 'U': '/', 'V': 'Q',\
                    'W': 'F', 'X': 'G', 'Y': 'K', 'Z': '6'}
    
    cryp_rus_arr = {'ю': 'W', 'п': 'Z', 'м': '2', 'ь': '3', 'я': 'S', 'з': '?',\
                    'ш': 'B', 'ж': 'J', 'х': '"', 'ф': 'X', 'е': 'H', 'ч': '8',\
                    'д': '9', 'р': 'L', 'в': '<', 'н': 'b', 'г': '$', 'э': 'I',\
                    'а': ',', 'у': 'T', 'с': 'a', 'о': '4', 'ё': '1', 'ъ': '5',\
                    'и': '(', 'ы': ';', 'б': 'E', 'ц': '#', 'й': '7', 'л': 'A',\
                    'т': ':', 'к': 'O', 'щ': '^', 'А': ')', 'Б': '@', 'В': "'",\
                    'Г': 'R', 'Д': 'M', 'Е': '=', 'Ё': 'C', 'Ж': '%', 'З': ']',\
                    'И': '+', 'Й': '-', 'К': 'U', 'Л': 'V', 'М': '/', 'Н': 'Q',\
                    'О': 'F', 'П': 'G', 'Р': 'K', 'С': '6', 'Т': '.', 'У': '!',\
                    'Ф': 'D', 'Х': '`', 'Ч': '_', 'Ш': '0', 'Щ': '\\', 'Ц': 'Y',\
                    'Ы': '&', 'Ъ': '*', 'Ь': '>', 'Э': 'P', 'Ю': 'N', 'Я': '['}
    
    total = ''
    
    if choice == 1:
        for i in message:
            if i == ' ':
                total += i
            
            else:
                total += cryp_rus_arr[i]

    elif choice == 2:
        for i in message:
            if i == ' ':
                total += i
                         
            else:
                total += cryp_eng_arr[i]
                
    else:
        print('Вы ввели неправильную команду.')
        
    return total
        
def decrypt(message, choice):
    decryp_eng_arr = {'W': 'a', 'Z': 'b', '2': 'c', '3': 'd', 'S': 'e', '?': 'f',\
                       'B': 'g', 'J': 'h', '"': 'i', 'X': 'j', 'H': 'k', '8': 'l',\
                       '9': 'm', 'L': 'n', '<': 'o', 'b': 'p', '$': 'q', 'I': 'r',\
                       ',': 's', 'T': 't', 'a': 'u', '4': 'v', '1': 'w', '5': 'x',\
                       '(': 'y', ';': 'z', 'E': 'A', '#': 'B', '7': 'C', 'A': 'D',\
                       ':': 'E', 'O': 'F', '^': 'G', ')': 'H', '@': 'I', "'": 'J',\
                       'R': 'K', 'M': 'L', '=': 'M', 'C': 'N', '%': 'O', ']': 'P',\
                       '+': 'Q', '-': 'R', 'U': 'S', 'V': 'T', '/': 'U', 'Q': 'V',\
                       'F': 'W', 'G': 'X', 'K': 'Y', '6': 'Z'}
    
    decryp_rus_arr = {'W': 'ю', 'Z': 'п', '2': 'м', '3': 'ь', 'S': 'я', '?': 'з',\
                       'B': 'ш', 'J': 'ж', '"': 'х', 'X': 'ф', 'H': 'е', '8': 'ч',\
                       '9': 'д', 'L': 'р', '<': 'в', 'b': 'н', '$': 'г', 'I': 'э',\
                       ',': 'а', 'T': 'у', 'a': 'с', '4': 'о', '1': 'ё', '5': 'ъ',\
                       '(': 'и', ';': 'ы', 'E': 'б', '#': 'ц', '7': 'й', 'A': 'л',\
                       ':': 'т', 'O': 'к', '^': 'щ', ')': 'А', '@': 'Б', "'": 'В',\
                       'R': 'Г', 'M': 'Д', '=': 'Е', 'C': 'Ё', '%': 'Ж', ']': 'З',\
                       '+': 'И', '-': 'Й', 'U': 'К', 'V': 'Л', '/': 'М', 'Q': 'Н',\
                       'F': 'О', 'G': 'П', 'K': 'Р', '6': 'С', '.': 'Т', '!': 'У',\
                       'D': 'Ф', '`': 'Х', '_': 'Ч', '0': 'Ш', '\\': 'Щ', 'Y': 'Ц',\
                       '&': 'Ы', '*': 'Ъ', '>': 'Ь', 'P': 'Э', 'N': 'Ю', '[': 'Я'}
    
    total = ''
         
    if choice == 1:
        for i in message:
            if i == ' ':
                total += i
                
            else:
                total += decryp_rus_arr[i]
    
    elif choice == 2:
        for i in message:
            if i == ' ':
                total += i
                
            else:
                total += decryp_eng_arr[i]                
    
    else:
        print('Вы ввели неправильную команду.')
        
    return total

def steg_crypt():
    print('Запущен шифровальщик')
    
    print('Выберите язык шифрования:\nРусский -> 1\nАнглийский -> 2')
    
    choice_num = int(input())
    
    message = input('Введите ваш текст: ')
    
    print('Если хотите ввести новый путь изображения наберите -> 1\
    \nЕсли хотите использовать путь готового изображения наберите -> 2')    
    
    choice_num_2 = int(input())
    
    if choice_num_2 == 1:
        img_path = input('Введите путь изображения: ')
    
    elif choice_num_2 == 2:
        img_path = 'sample.jpg'
        
    else:
        print('Вы ввели неправильную команду.')
        
    img = Image.open(img_path)
    draw = ImageDraw.Draw(img)
    
    width = img.size[0]
    height = img.size[1]
    
    if os.path.exists('save'):
        
        os.remove('save\enc_img.png')
        os.remove('save\keys.txt')
        os.rmdir('save')
    
    os.mkdir('save')
    
    pix = img.load()
    f = open('save\keys.txt', 'w')
    crypt_message = crypt(message, choice_num)
    
    print(f'Ваше сообщение зашивровано как: {crypt_message}')
    
    for elem in ([ord(elem) for elem in crypt_message]):
        
        key = (randint(1, width - 10), randint(1, height - 10))
        g, b = pix[key][1:3]
        draw.point(key, (elem, g, b))
        f.write(str(key)+ '\n')
        
    print('Ключи сохранены в save\keys.txt')
    print('Зашифрованное изображение сохранен в save\enc_img.png')
    
    img.save('save\enc_img.png', 'PNG')
    
    print('Открывается зашифрованное изображение...')
    os.startfile('save\enc_img.png')
    
    f.close()
    
    return choice_num
    
def steg_decrypt(choice):
    print('Запущен дешифровальщик')
    
    temp_a = []
    keys_a = []
    
    im_path = input('Введите путь зашифрованного изображения:')
    img = Image.open(im_path)
    
    pix = img.load()
    f = open('save\keys.txt', 'r')
    y = str([line.strip() for line in f])
    
    for i in range(len(findall(r'\((\d+)\,', y))):
        keys_a.append((int(findall(r'\((\d+)\,', y)[i]),\
                           int(findall(r'\,\s(\d+)\)', y)[i])))
    
    for key in keys_a:
        temp_a.append(pix[tuple(key)][0])
    
    crypt_message = ''.join([chr(elem) for elem in temp_a])
    decrypt_message = decrypt(crypt_message, choice)

    print(f'Сообщение: {crypt_message}, расшифровано как: {decrypt_message}')
    return decrypt_message

def main():
    choice = steg_crypt()
    message = steg_decrypt(choice)
    
if __name__ == '__main__':
    main()
```
